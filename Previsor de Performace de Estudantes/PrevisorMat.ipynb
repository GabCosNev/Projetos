## Dataset e Transformações

```python
data = pd.read_csv('student-mat.csv', delimiter = ';')
data
```

```python
data['Media'] = data.iloc[:, -3:].mean(axis=1).round()
data
```

```python
data['Media'] = data['Media'].apply(meanCalculator.categorizar_performance)
data
```

```python
X = data.iloc[:, [0 ,1 ,2 ,3 ,12 ,13 ,14 ,15, 16, 17, 18, 20, 21, 28, 29]].values
y = data.iloc[:, -1].values
```

```python
pd.DataFrame(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
lb = LabelEncoder()

for column in [0, 1, 3, 7, 8, 9, 10, 11, 12]:
    X_train[:, column] = lb.fit_transform(X_train[:, column])
    X_test[:, column] = lb.transform(X_test[:, column]) 
```

```python
pd.DataFrame(X_train)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
X_train[:, [2, 5, 13]] = scaler.fit_transform(X_train[:, [2, 5, 13]])
X_test[:, [2, 5, 13]] = scaler.transform(X_test[:, [2, 5, 13]])
```
